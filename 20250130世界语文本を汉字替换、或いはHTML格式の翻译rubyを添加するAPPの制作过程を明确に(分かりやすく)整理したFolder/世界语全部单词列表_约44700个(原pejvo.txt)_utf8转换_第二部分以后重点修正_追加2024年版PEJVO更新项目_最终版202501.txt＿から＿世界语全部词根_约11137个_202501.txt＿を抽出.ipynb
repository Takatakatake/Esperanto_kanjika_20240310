```python
# 设置包含世界语单词和意义的输入文本文件路径。(エスペラントの単語と意味が書かれたテキストファイルのパスを設定します。)
input_file = "一时的_PEJVO(世界语全部单词列表)を小文字・字上符形式に转换.txt"

# 创建一个空集合来保存词根。 (語根を保存するための空の集合を作成します。)
roots = set()

with open(input_file, 'r', encoding='utf-8') as file:
    # 遍历每一行。 (各行をループします。)
    for line in file:
        line=line.replace('-','/')##20240618追加
        # 提取":"之前的部分。(":"が出てくるまでの部分を取り出します。)
        word = line.split(":")[0]
        word = word.lstrip('/')
        # さらに取り出した部分を'-'、' '、','で分ける。","もごくまれに存在する('tial')
        parts = re.split('-| |,', word)##  parts = re.split('| |,', word)は「何もない文字列」で分割してしまうことになるので注意。
        part=parts[0]

        # 部分を"/"と" "でさらに分割します。
        subparts = re.split('/| ', part)
        # 各サブパートを語根として追加します。
        for subpart in subparts:
            x=subpart.strip()
            if len(x)>=2:#长度小于等于2的词根实在是不必要 (長さが2以下の語根は流石に不要)
                roots.add(x)

##去除重复 在转换为字上符格式后进行,否则会出现重复?  (重複を削除　字上符形式に変換したあとでないと、重複が生じてしまう？)
unique_list = list(set(roots))

# 按词根长度的倒序(从长到短)对词根进行排序。  (語根を文字数順の逆に(文字数の多いものから順に)並び替えます。)
# 对于相同长度的词根,按字母顺序排序。 (同じ文字数の語根についてはアルファベット順に並び替えます。)
sorted_roots = sorted(unique_list, key=lambda root: (-len(root), root))

# 将结果写入新的文本文件。 (結果を新しいテキストファイルに書き出します。)
with open('世界语全部词根_约11137个_202501.txt', 'w', encoding='utf-8') as file:
    for root in sorted_roots:
        file.write(root + "\n")
```

```python
##检查是否成功 (うまくいったかをチェック)
###确认是否有重复行 (重複した行がないかを確認)
def check_duplicate_lines(file_path):
    unique_lines = set()
    duplicate_lines = set()

    with open(file_path, 'r') as file:
        for line in file:
            if line in unique_lines:
                duplicate_lines.add(line)
            else:
                unique_lines.add(line)

    return duplicate_lines

#  指定文本文件路径并检测重复行  (テキストファイルのパスを指定して重複行を検出)
file_path = '世界语全部词根_约11137个_202501.txt'  # 请替换为实际的文件路径  (実際のファイルパスに置き換えてください)
duplicates = check_duplicate_lines(file_path)

if duplicates:
    print('存在重复行:')
    for line in duplicates:
        print(line)
else:
    print('没有重复行')
```
